 # **Outpur using TTS model**

```python
from transformers import pipeline
from IPython.display import Audio, display

full_text = "Hi user! Here is what I can see in the image: " + gemini_text

narrator = pipeline("text-to-speech", model="suno/bark-small")
audio_out = narrator(full_text)

audio_arr = audio_out["audio"]
sr = audio_out["sampling_rate"]

display(Audio(audio_arr, rate=sr))
```
